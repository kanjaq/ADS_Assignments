```python
#Display the top 5 video games by global sale
Top_global = df.nlargest(5, 'Global_Sales')
print(Top_global)
```

```python
#Calculate the total sales in North America for the 'Action' genre.
North_sales = df[df['Genre']=='Action']['NA_Sales'].sum()
print(North_sales)
```

```python
#Find the top 3 publishers by global sales.
Top_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(3)
print(Top_publishers)
```

```python
#Calculate the average sales in Europe for the 'Shooter' genre.
Avg_Europe_sales = df[df['Genre']=='shooter']['EU_Sales'].mean()
print(Avg_Europe_sales)
```

```python
#Find the game with the highest sales in Japan in the 'Sports' genre.
highest_sales = df[df['Genre']]['EU_Sales'].max()
print(highest_sales)
```
